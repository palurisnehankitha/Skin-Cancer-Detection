```python
import pandas as pd
import joblib
```

```python
model = joblib.load("skin_cancer_model.pkl")

df = pd.read_csv("hmnist_28_28_RGB.csv")

sample = df.iloc[0:1, :-1]

actual = df.iloc[0, -1]
```

```python
prediction = model.predict(sample)

print("Actual Label    :", actual)
print("Predicted Label :", prediction[0])
```

```python
row_number = 150

sample = df.iloc[row_number:row_number+1, :-1]

prediction = model.predict(sample)

print("Predicted Class :", prediction[0])
```

```python
classes = {
    0: "Melanocytic Nevi",
    1: "Melanoma",
    2: "Benign Keratosis",
    3: "Basal Cell Carcinoma",
    4: "Actinic Keratoses",
    5: "Vascular Lesions",
    6: "Dermatofibroma"
}

print("Disease :", classes[prediction[0]])
```
